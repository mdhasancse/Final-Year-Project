```python
import pandas as pd
path='Titanic.csv'
df=pd.read_csv(path)
df
```

```python
df.describe()
```

```python
df.Age= df.Age.fillna(df.Age.median())
df.describe()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
le_Embarked=LabelEncoder()
df['Sex_n']=le_Sex.fit_transform(df['Sex'])
#df['Embarked_n']=le_Embarked.fit_transform(df['Embarked'])
df.head()
```

```python
df['Embarked'].value_counts()
```

```python
df.Embarked= df.Embarked.fillna('S')
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].unique()
```

```python
df.head()
```

```python
df['Embarked_n']=le_Embarked.fit_transform(df['Embarked'])
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(df.Survived)
plt.show()
```

```python
x=df.drop(['Survived','PassengerId','Name','Sex','Ticket','Cabin','Embarked'],axis=1)
y=df.Survived
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(x,y)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,y_pred)
score=(cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
print('Accuracy of this model is:',score*100)
```

```python
pred=int(model.predict([[3,35.0,0,0,8.0500,1,2]]))
if pred==0:
    print('Passenger not survived')
else:
    print('Passenger survived')
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', max_iter=1000)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,y_pred)
score=(cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
print('Accuracy of this model is:',score*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators = 100) 
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm=confusion_matrix(y_test,y_pred)
score=(cm[0][0]+cm[1][1])/(cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1])
print('Accuracy of this model is:',score*100)
```
